# Introduction to Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
def draw_pie(data,labels):
    #define Seaborn color palette to use
    colors = sns.color_palette('pastel')[0:5]
    #create pie chart
    plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
    plt.show()
```

## Total_bill

```python
sns.set_theme()
```

```python
sns.histplot(data = tips_df, x = 'total_bill' ,hue = 'sex');
```

```python
sns.displot(data=tips_df, x= 'total_bill', col= 'time'  ,kde= True, hue = 'sex');
```

```python
sns.displot(data=tips_df, x= 'total_bill', col= 'day'  ,kde= True, hue='sex');
```

```python
sns.displot(data=tips_df, x= 'total_bill', col= 'sex'  ,kde= True);
```

```python
sns.boxplot(data = tips_df, x ='time',  y='total_bill' , hue ='sex');
```

```python
sns.boxplot(data = tips_df, x = 'day', y ='total_bill', hue = 'sex');
```

```python
sns.displot(data =tips_df, x ='total_bill', col ='smoker', kde= True );
```

## Day

```python
sns.countplot(data = tips_df, x = 'day');
```

```python
data = list(tips_df['day'].value_counts())
labels = list(tips_df['day'].value_counts().index)
draw_pie(data,labels)
```

```python
sns.countplot(data = tips_df, x = 'day', hue='time');
```

## Sex

```python
sns.countplot(data =tips_df, x ='sex');
```

```python
data = list(tips_df['sex'].value_counts())
labels = list(tips_df['sex'].value_counts().index)
draw_pie(data,labels)
```

```python
sns.barplot(data = tips_df, x = 'sex', y = 'tip', estimator=np.mean);
```

```python
sns.countplot(data = tips_df, x = 'sex', hue ='time');
```

## Smoke
